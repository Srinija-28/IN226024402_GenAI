LLM

```python
from langchain_community.llms import HuggingFacePipeline
from transformers import pipeline

pipe = pipeline("text-generation", model="distilgpt2", max_new_tokens=50)

llm = HuggingFacePipeline(pipeline=pipe)

print(llm.invoke("What is LangChain?").strip())
```

Prompt

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate.from_template("Explain {topic} in simple terms")

print(prompt.format(topic="AI"))
```

Chain

```python
from langchain_core.output_parsers import StrOutputParser

chain = prompt | llm | StrOutputParser()

print(chain.invoke({"topic": "Machine Learning"}).strip())
```

Memory

```python
chat_history = []

user1 = "Hi"
chat_history.append(user1)

response1 = llm.invoke(user1)
chat_history.append(response1)

print("AI:", response1.strip())

user2 = "What did I just say?"
chat_history.append(user2)

context = " ".join(chat_history)

response2 = llm.invoke(context)

print("AI:", response2.strip())
```

Tool + Agent

```python
def calculator(q):
    return str(eval(q))

print("Tool Example:", calculator("5*10"))
```

Vector Store

```python
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings

# ✅ Use small model (IMPORTANT FIX)
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

texts = [
    "LangChain is powerful",
    "AI is the future"
]

db = FAISS.from_texts(texts, embeddings)

results = db.similarity_search("AI")

for r in results:
    print(r.page_content)
```
